```python
# cleaning the data
cleaned_data = data.copy()

# removing null values
cleaned_data.dropna(subset=['29. State the 3 main challenges facing your county government.'], inplace=True)
cleaned_data['29. State the 3 main challenges facing your county government.'].isna().sum()
```

```python
# Removing null values
target_sentences = list(cleaned_data.columns)[1:]
target_sentences
```

```python
# Concating function for challenges
def concating_func(x):
    if '\n' in x:
        challenges = x.split('\n')
        return [challenge.strip() for challenge in challenges if challenge.strip() != '']
    else:
        return [x]


cleaned_data['29. State the 3 main challenges facing your county government.'] = cleaned_data['29. State the 3 main challenges facing your county government.'].apply(concating_func)
# Print the resulting DataFrame
cleaned_data['29. State the 3 main challenges facing your county government.'].head()
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')


def tokenize_text(list):
  results = []
  for text in list:
    tokens = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=128)
    results.append(tokens)
  return tokens

cleaned_data.insert(1, 'tokenized text', cleaned_data['29. State the 3 main challenges facing your county government.'].apply(tokenize_text))
```

```python
def tuner(number):
  if number < 2.5:
    return 0
  elif number >= 2.5 and number < 2.55:
    return 1
  elif number >= 2.6:
    return 3
  else:
    return 2

def myPredictor(query_tokens):
  with torch.no_grad():
    query_outputs = model(**query_tokens)
    query_embedding = query_outputs.last_hidden_state.mean(dim=1)
    min_range=1
    max_range=3

    target_outputs = model(**encoded_target_sentence)
    target_embeddings = target_outputs.last_hidden_state.mean(dim=1)


    similarity_scores = cosine_similarity(query_embedding, target_embeddings)
    scaled_similarity_scores = [((score + 1) / 2) * (max_range - min_range) + min_range for score in similarity_scores][0]
    tuned_scores = [tuner(score) for score in scaled_similarity_scores]


    return tuned_scores
```
